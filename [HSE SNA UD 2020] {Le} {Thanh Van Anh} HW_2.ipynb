````python
# Task 2.3
url = "https://vk.com/al_wall.php" # Choose API to get Vk wall posts of HSEISA
headers = {
    'authority': 'vk.com',
    'pragma': 'no-cache',
    'cache-control': 'no-cache',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36',
    'x-requested-with': 'XMLHttpRequest',
    'content-type': 'application/x-www-form-urlencoded',
    'accept': '*/*',
    'origin': 'https://vk.com',
    'sec-fetch-site': 'same-origin',
    'sec-fetch-mode': 'cors',
    'sec-fetch-dest': 'empty',
    'referer': 'https://vk.com/hseisa',
    'accept-language': 'en-US,en;q=0.9,vi;q=0.8',
    'cookie': 'remixlang=3; remixlhk=28a84c8667e3a93441; remixscreen_depth=24; remixdt=14400; remixstid=21374603_1aNm3I8dGw0wGRwYJZ2owLHIE6fNA8L8zmSA81z3Jt0; remixflash=0.0.0; remixscreen_width=1536; remixscreen_height=864; remixscreen_dpr=1.25; remixscreen_orient=1; tmr_lvid=2039fca8a154d7220073224256c9d617; tmr_lvidTS=1577459759327; remixusid=NzZiYmUxNzMxMGUyMWI2NjYxZWVlOGMx; remixseenads=0; remixab=1; remixsid=ca1e2a68196a0a432a68813787666d3fc1b6b7e74ac94804694e4418be759; remixrefkey=9a91aac2149cb8e1bb; remixscreen_winzoom=1; remixgp=a8c162513dd6165b2e7879c8e5f16482; tmr_detect=1%7C1607525380599; tmr_reqNum=105'
}
params = {
    'act': 'get_wall',
    'al': 1,
    'fixed':1646,
    'offset': 10,
    'onlyCache': False,
    'owner_id': -19868623,
    'type': 'own',
    'wall_start_from': 0
}

def getResultPostContent(url, headers, params): # create function request and return essential data
    resultPostContents = []
    try:
        response = requests.request("POST", url, headers=headers, params=params) # call request to receive data
        if (response.json() and response.json()['payload']): # check if request return correct data
            dataText = response.json()['payload'][1][0] # get html data from response
            dataSoup = BeautifulSoup(dataText, "html.parser") # format html data with beautiful soup
            postContent = dataSoup.find_all('div', { 'class': 'post_content'}) # get all post content 

            for post in postContent: # iterate to get wall_post, like, comment, ...
                wall_post = post.find('div', { 'class': 'wall_text' }).get_text()
                like = '0' # get number of likes
                if post.find('a', { 'class': '_like' }): #find like box container
                    like = post.find('a', { 'class': '_like' }).find('div', { 'class': 'like_button_count' }).get_text() or '0'
                comment = '0' # get number of comment
                if post.find('a', { 'class': '_comment' }): #find comment box container
                    comment = post.find('a', { 'class': '_comment' }).find('div', { 'class': 'like_button_count' }).get_text() or '0'
                share = '0' # get number of share
                if post.find('a', { 'class': '_share' }): # find share box container
                    share = post.find('a', { 'class': '_share' }).find('div', { 'class': 'like_button_count' }).get_text() or '0'
                view = post.find('div', { 'class': '_views' }).get_text() or '0'
                resultPostContents.append({
                    'wall_post': wall_post,
                    'like': like,
                    'comment': comment,
                    'share': share,
                    'view': view
                })
        return resultPostContents # return if request is blocked or no more data returned
    except:
        return resultPostContents
def printContent(content): # create function to display data
    print('\n#######################')
    print(content['wall_post'])
    print('Number of like: ' + content['like'])
    print('Number of comment: ' + content['comment'])
    print('Number of share: ' + content['share'])
    print('Number of view: ' + content['view'])

# get result content from url
result = getResultPostContent(url, headers, params)
for content in result:
    printContent(content)
````
