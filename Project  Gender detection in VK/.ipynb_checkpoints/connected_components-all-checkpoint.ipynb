```python
# Preliminary imports

import urllib
import networkx as nx
from networkx.algorithms import community
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
from tqdm.notebook import tqdm
%matplotlib inline
import csv
import requests

from networkx.algorithms.community import greedy_modularity_communities
from networkx.algorithms.community import label_propagation_communities
```

```python
token = "846137628461376284613762a08414a9d78846184613762db8256752edc2e1df41d5ebb"
group_id = 19868623
interesting_fields = "sex,bdate,country,education,relation"
```

```python
def getGroupMembers():
    group_members = f"https://api.vk.com/method/groups.getMembers?group_id={group_id}&fields={interesting_fields}" +\
           f"&access_token={token}&v=5.126"
    json_response = requests.get(group_members).json()
    if json_response.get('error'):
        return []
    return json_response['response']['items']
```

```python
list_members = getGroupMembers()
list_id_member=[d['id'] for d in list_members] 
```

```python
print(len(list_id_member))
```

```python
def getUserEdges():
    edges = []
    with open('user_data_all.pickle', 'rb') as f:
        list_friends = pickle.load(f)
        f.close()
    list_friends_keys = list_friends.keys()
    with open('user_connections_all.pickle', 'rb') as s:
        list_connection = pickle.load(s)
        s.close()
    for item in list_connection:
        if (item['name'] in list_friends_keys and item['friend_id'] in list_friends_keys):
            edges.append((item['name'], item['friend_id']))
    return edges
```

```python
#Loading the VK network frame
edges = getUserEdges()

G = nx.Graph()
G.add_edges_from(edges)

print("Number of nodes in a graph: ", len(G.nodes()))
print("Number of edges in a graph: ", len(G.edges()))
```

```python
plt.figure(figsize = (15,10))
nx.draw(G, node_size = 10, edges_width = .1)
```

### Find connected components

```python
sorted_subgraphs = sorted(nx.connected_components(G), key=len, reverse=True)
```

```python
len(sorted_subgraphs[0])
```

```python
l = []
for item in sorted_subgraphs:
    l.append(len(item))
_ = plt.hist(l, bins = 10, log = True)
```

### Largest connected component

```python
G_subgraph = G.subgraph(sorted_subgraphs[0])
```

```python
plt.figure(figsize = (15,10))
nx.draw(G_subgraph, node_size = 100, edges_width = .1)
```

### Save subgraph

```python
nx.write_edgelist(G_subgraph, "G_subgraph_all.gz")
```

### Get gender info about user in largest CC

```python
with open("user_data_all.pickle", 'rb') as f:
    users_info = pickle.load(f)
```

```python
users_info
```

```python
male = set()
male_indicator = []
idx_indicator = []
```

```python
for user_id in sorted_subgraphs[0]:

    if user_id not in list_id_member:
        idx_indicator.append(1)
    else:
        idx_indicator.append(0)
    male_indicator.append(0)
    if user_id in users_info.keys():   
        if "sex" in users_info[user_id]:
            if users_info[user_id]["sex"] == 2:
                male.add(user_id) 
                male_indicator[-1] = 1
```

```python
list_connections_largest = []
list_friends_largest = []

with open('user_data_all.pickle', 'rb') as f:
    list_friends = pickle.load(f)
    list_friends_keys = list_friends.keys()
    for user_id in sorted_subgraphs[0]:
        if user_id in list_friends_keys:
            list_friends_largest.append(list_friends[user_id])
    f.close()
with open('user_connections_all.pickle', 'rb') as s:
    list_connection = pickle.load(s)
    for item in list_connection:
        if item['name'] in sorted_subgraphs[0] and item['friend_id'] in sorted_subgraphs[0]:
            list_connections_largest.append(item)
    s.close()
with open('user_data_largest_all.pickle', 'wb') as f:
    pickle.dump(list_friends_largest, f)
    f.close()
with open('user_connections_largest_all.pickle', 'wb') as f:
    pickle.dump(list_connections_largest, f)
    f.close()
```
