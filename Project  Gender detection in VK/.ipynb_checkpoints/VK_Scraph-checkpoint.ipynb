```python
import requests
from tqdm import tqdm
import pandas as pd
import pickle
import time
```

```python
token = "846137628461376284613762a08414a9d78846184613762db8256752edc2e1df41d5ebb"
group_id = 19868623
```

```python
interesting_fields = "sex,bdate,country,education,relation"
```

```python
def getGroupMembers():
    group_members = f"https://api.vk.com/method/groups.getMembers?group_id={group_id}&fields={interesting_fields}" +\
           f"&access_token={token}&v=5.126"
    json_response = requests.get(group_members).json()
    if json_response.get('error'):
        return []
    return json_response['response']['items']
```

```python
list_members = getGroupMembers()
print(list_members)
```

```python
def process_person_info(person):
    person_info = {}
    person_info["first_name"] = person["first_name"]
    person_info["last_name"] = person["last_name"]
    person_info["id"] = person["id"]
    if ("country" in person.keys() and person["country"]["title"]):
        person_info["country"] = person["country"]["title"]
    person_info["sex"] = person["sex"]
    if ("university_name" in person.keys()):
        person_info["university_name"] = person["university_name"]
    return person_info
```

```python
def getFriendsByUserId(user_id):
    member_ids = f"https://api.vk.com/method/friends.get?user_id={user_id}&fields={interesting_fields}" +\
           f"&access_token={token}&v=5.126"
    json_response = requests.get(member_ids).json()
    if json_response.get('error'):
        return []
    return json_response['response']['items']

def getAllFriends():
    friends = dict()
    connections = []
    for member in tqdm(list_members):
        if member['id'] not in friends.keys():
            friends[member['id']] = process_person_info(member)
        users = getFriendsByUserId(member['id'])
        for user in users:
            if user['id'] not in friends.keys():
                connections.append({'friend_id': user['id'], 'name': member['id']})
                friends[user['id']] = process_person_info(user)
    return (friends, connections)
```

```python
(list_friends, list_connections) = getAllFriends()
```

```python
with open("male_all.txt", 'r') as f:
    user_ids = f.read().split()
    
with open("female_all.txt", 'r') as f:
    user_ids.extend(f.read().split())
```

```python
from tqdm import tqdm
```

```python
res = []
def get_user_url(user_ids):
    # see more about options : hhttps://vk.com/dev/users.get

    return f"https://api.vk.com/method/users.get?user_ids={','.join(user_ids)}&fields={fields}" +\
           f"&access_token={token}&v=5.126"

for i in tqdm(range(len(user_ids))):
    r = requests.get(get_user_url([user_ids[i]])).json()
    res.extend(r["response"])
```

```python
def transform_dict(cur_dict):
    result_dict = dict()
    for field_name, field_data in cur_dict.items():
        if type(field_data) == dict:
            for value_key, value_data in field_data.items():
                result_dict[field_name + "_" + value_key] = value_data
        else:
            result_dict[field_name] = field_data
    return result_dict
```

```python
res_dict = dict()
for item in res:
    res_dict[item["id"]] = transform_dict(item)
```
